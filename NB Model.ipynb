```python
import pandas as pd
import numpy as np
import ast

from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
```

```python
df = pd.read_csv("recommendations_df.csv")
```

```python
#convert the new datafrem columns to strings
for col in df:
    df[col] = df[col].astype("string")
```

```python
# Function to parse the string representation of lists and concatenate the first three items
def get_first_three_items_concatenated(items):
    # Convert string representation of list to an actual list
    items_list = ast.literal_eval(items)
    # Return the first three items concatenated as a string
    return ', '.join(items_list[:3])

# Applying the function to the 'recommended_items' column
df['first_three_items'] = df['recommended_items'].apply(get_first_three_items_concatenated)

# Display the DataFrame
print(df[['CustomerID', 'first_three_items']])
```

```python
data = pd.read_csv("superstore.csv")
```

```python
# create rating column

start = 1
end = 10
width = end - start
rating = round((data['Quantity'] - data['Quantity'].min())/(data['Quantity'].max() - data['Quantity'].min()) * width + start).astype(int)
data['Rating'] = rating
```

```python
group = data.groupby('Customer.ID')

df2 = group.apply(lambda x: x['Rating'].unique().tolist()) 
df2.values.tolist()
```

```python
interaction_matrix = data.pivot_table(index='Customer.ID',columns='Product.Name',values='Rating').fillna(0)

from sklearn.model_selection import train_test_split

# Train-test split (optional, but recommended for evaluation)
user_features_train, user_features_test, ratings_train, ratings_test = train_test_split(
    interaction_matrix, df2.values.tolist(), test_size=0.3, random_state=42)
```

```python

# Function to evaluate recommendations
def evaluate_recommendations(df, actual_df):
    y_true = []
    y_pred = []
    
    for _, row in df.iterrows():
        actual_items = df[df['CustomerID'] == row['CustomerID']]['first_three_items'].tolist()
        recommended_items = row['recommended_items']
        
        for item in actual_items:
            y_true.append(1 if item in recommended_items else 0)
            y_pred.append(1 if item in recommended_items else 0)
        
        for item in recommended_items:
            if item not in actual_items:
                y_true.append(0)
                y_pred.append(1)
    
    accuracy = accuracy_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred, zero_division=0)
    recall = recall_score(y_true, y_pred, zero_division=0)
    
    return accuracy, precision, recall

# Get actual items from the test set
actual_df = df[df['CustomerID'].isin(user_features_test.index)]

# Evaluate recommendations
accuracy, precision, recall = evaluate_recommendations(df, actual_df)

print(f"Accuracy: {accuracy*100:.2f}")
print(f"Precision: {precision*100:.2f}")
print(f"Recall: {recall*100:.2f}")
```
